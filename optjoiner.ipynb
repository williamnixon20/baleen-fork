```python
from BCacheSim import cachesim
# import file utils from cachesim
from BCacheSim.cachesim import utils

decisions = {}
for i in range(0,7):
    path_day = f"/home/cc/Baleen-FAST24/tmp/optday{i}/example/201910_Region1_0_0.1/decisions_utility_service_time_size_fixed_ea_5892.86.pkl.bz"
        
    data = utils.compress_load(path_day)
    for key, value in data.items():
        if key not in decisions:
            decisions[key] = []
        for val in value:
            decisions[key].append(val)
    
decisions = {k: tuple(v) for k, v in decisions.items()}
```
