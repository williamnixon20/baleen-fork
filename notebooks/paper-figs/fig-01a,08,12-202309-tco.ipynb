```python
dfs_01 = get_data(dfc_raw, default_sample_ratio=.1)
```

```python
params =  {
        "DWPDClose": True,
        "PracticalAP": True,
        "ShortLabel": [
            "CoinFlip",
            "RejectX",
            "Baleen",
            "Baleen (No Prefetch)",
            # "Baleen (All on Partial Hit)",
            # "Baleen (ML-Range on Partial Hit)",
            # "Baleen (ML Prefetch)",
        ],
        #3.75,
        "Target Cache Size": ef.DEFAULT_CSIZE,
        "Target DWPD": [3.75, 7.5, 15.0, 20.0,17.5, 10.0, 12.5], # 3, 3.75,
    #1,3
}
df_01 = filter_df(dfs_01["exp"], params)
```

```python
df_01f = fillright(sampleright(df_01, idx=['ShortLabel', 'Target Write Rate', 'Target DWPD']), idx=['ShortLabel', 'Target Write Rate', 'Target DWPD'], verbose=1)
```

```python
y = 'P100ServiceTimeUtil@10m'
cost_dwpd = 1/7.5 * 170
cost_dtutil = 281 * 36

dff = df_01f


dx_nopf = filter_df(dff, {'ShortLabel': 'Baleen (No Prefetch)', 'Target DWPD': 7.5}).set_index("Region")
dx_ = filter_df(dff, {'ShortLabel': 'Baleen'})
dx_cf = filter_df(dff, {'ShortLabel': 'CoinFlip', 'Target DWPD': 7.5}).set_index("Region")
dx_o = filter_df(dff, {'ShortLabel': 'RejectX', 'Target DWPD': 7.5}).set_index("Region")
dx_o["Cost (flash writes)"] = dx_o["Target DWPD"] * cost_dwpd
dx_o["Cost (reads)"] =  cost_dtutil
dx_o["TCO (ops)"] = dx_o["Cost (flash writes)"] + dx_o["Cost (reads)"]
dx_cf["Cost (flash writes)"] = dx_cf["Target DWPD"] * cost_dwpd
dx_cf["Cost (reads)"] =  dx_cf["P100ServiceTimeUtil@10m"] /  dx_o["P100ServiceTimeUtil@10m"] * cost_dtutil
dx_cf["TCO (ops)"] = dx_cf["Cost (flash writes)"] + dx_cf["Cost (reads)"]
dx_nopf["Cost (flash writes)"] = dx_nopf["Target DWPD"] * cost_dwpd
dx_nopf["Cost (reads)"] =  dx_nopf["P100ServiceTimeUtil@10m"] /  dx_o["P100ServiceTimeUtil@10m"] * cost_dtutil
dx_nopf["TCO (ops)"] = dx_nopf["Cost (flash writes)"] + dx_nopf["Cost (reads)"]

ax = plt.gca()
rows = []
bestrows = []
dx = dx_.sort_values("RegionLabel")
for regionlabel, dx__ in dx_.groupby("RegionLabel"):
    region = dx__["Region"].values[0]
    dx__["Cost (flash writes)"] = dx__["Target DWPD"] * cost_dwpd
    dx__["Cost (reads)"] = dx__["P100ServiceTimeUtil@10m"] / dx_o["P100ServiceTimeUtil@10m"].loc[region] * cost_dtutil
    dx__["TCO (ops)"] = dx__["Cost (flash writes)"] + dx__["Cost (reads)"]
    dx__["%-Cost (reads)"] = dx__["Cost (reads)"] / dx__["TCO (ops)"] * 100
    dx__["%-Cost (flash writes)"] = dx__["Cost (flash writes)"] / dx__["TCO (ops)"] *100
    dx__["P100ServiceTimeUtil@10m-NormNoCache"] = dx__["P100ServiceTimeUtil@10m"] / dx_o["P100ServiceTimeUtil@10m"].loc[region] * 100
    bestr = dx__.iloc[dx__["TCO (ops)"].argmin()]
    tco = {
        'Region': region,
        'RegionLabel': maps.region_labels[region],
        'Baleen': dx__.set_index("Target DWPD").loc[7.5]["TCO (ops)"],
        'Baleen-TCO': bestr["TCO (ops)"],
        'Baleen-TCO (disk reads)': bestr["Cost (reads)"],
        'Baleen-TCO (flash writes)': bestr["Cost (flash writes)"],
        'Baleen (No Prefetch)': dx_nopf["TCO (ops)"].loc[region],
        'RejectX': dx_o["TCO (ops)"].loc[region],
        'CoinFlip': dx_cf["TCO (ops)"].loc[region],
    }
    bestrows.append(bestr)
    dx__.plot(x="Target DWPD", y=y, label=maps.region_labels[region], ax=ax, marker='.')
    plt.scatter([bestr['Target DWPD']], [bestr[y]], marker='x')
    rows.append(tco)
dxtt = pd.DataFrame(rows)

plt.ylim(0, 60)
plt.xlim(7.5, 20)
plt.xlabel("Target DWPD")
plt.tick_params(axis='y', which='both', labelleft=False)
plt.ylabel("Peak Backend Load (%)")#\n(% of no cache)")
plt.legend(ncols=3)
savefig("tco", f"dwpd_peakload")
```
